### Apriori Preparation

Before converting dataset to transaction view, we should convert pandas-data to list-data and then list-data to numpy-data.

```python
arr = []
for i in range(df.shape[0]):
    count = int((i/7500)*100)
    if (i % 500 == 0):
        print(count, '% -- Loading')  
        
    arr.append([str(df.values[i,j]) for j in range(df.shape[1])])
       
arr = np.array(arr)
print(arr.shape)
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder


te = TransactionEncoder()
data = te.fit_transform(arr) #Apply one-hot-encoding on our dataset
data = pd.DataFrame(data, columns = te.columns_) #Creating a new DataFrame from our Numpy array
print(data.shape)  
```

```python
data.head(3)
```

Support Code

```python
from mlxtend.frequent_patterns import apriori
freq_rules = apriori(data, min_support = 0.01, use_colnames = True) #Support Code
freq_rules
```

Confidence code

```python
from mlxtend.frequent_patterns import association_rules # Confidence code
association_rules(freq_rules,metric="confidence",min_threshold=0.01) # associate itemsets.
```

Lift code

```python
from mlxtend.frequent_patterns import association_rules # Lift code
association_rules(freq_rules,metric="lift",min_threshold=0.01)
```
